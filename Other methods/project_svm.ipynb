```python
import pandas as pd
import numpy as np

#df_train = pd.read_csv('fTrain.csv')
df_train = pd.read_csv('fTrainNorm.csv')

df_train.head()
```

```python
from sklearn.model_selection import train_test_split

y_train = df_train[['PAX']]

df_train.drop(df_train.columns[[1,2,5,6,12,13,14]], axis=1, inplace=True)

df_train.head()
```

```python
# For submitting

#df_test = pd.read_csv('fTest.csv')
df_test = pd.read_csv('fTestNorm.csv')
df_test.drop(df_test.columns[[1,2,5,6,12,13,14]], axis=1, inplace=True)

X_train = df_train
Y_train = y_train
X_test = df_test

Y_train = np.ravel(Y_train)
```

```python
df_test.head()
```

```python
#For testing here

X_train, X_test, Y_train, Y_test = train_test_split(df_train, y_train, test_size=0.3, random_state=42)
Y_train = np.ravel(Y_train)
Y_test  = np.ravel(Y_test)
```

```python
# SVM
import numpy as np
from __future__ import print_function

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn import metrics
from sklearn.cross_validation import cross_val_score, cross_val_predict

#params_grid = {'C': [0.001, 0.005, 0.01, 0.1, 0.2, 0.5, 0.7, 1, 2, 10, 100],
#               'gamma': [0.0001, 0.001, 0.005, 0.01, 0.5, 0.1],
#               'degree': [1, 2, 3],
#               'kernel': ['linear','rbf', 'poly', 'sigmoid'] }

#grid_clf = GridSearchCV(SVC(class_weight='balanced'), params_grid, cv=10)
#grid_clf.fit(X_train, y_train)

#clf = SVC(gamma='auto', verbose=True)
#clf.fit(X_train, y_train) 

###clf = SVC(gamma='auto')
###scores = cross_val_score(clf, X_train, Y_train, cv=10, verbose=1,n_jobs=8)
###scores
###clf.fit(X_train, Y_train)


C_range = np.logspace(0.01, 10, 15)
gamma_range = np.logspace(0.001, 10, 15)
kernel_range = ['rbf', 'sigmoid']
param_grid = dict(gamma=gamma_range, C=C_range, kernel=kernel_range)
cv = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
clf = GridSearchCV(SVC(verbose=True), param_grid=param_grid, cv=cv, n_jobs=8, verbose=1)
clf.fit(X_train, Y_train)
```

```python
Y_pred = clf.predict(X_test)
Y_pred.shape[0]
```

```python
# Write file

import csv
with open('y_pred.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(['Id', 'Label'])
    for i in range(Y_pred.shape[0]):
        #writer.writerow([i, int(Y_pred[0][i])])
        writer.writerow([i, Y_pred[i]])
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", 100*accuracy_score(Y_test, Y_pred))
```
